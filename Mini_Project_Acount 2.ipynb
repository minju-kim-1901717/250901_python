##### ë¬¸ì œ 5.1: ë°ì½”ë ˆì´í„° êµ¬í˜„

    - í•¨ìˆ˜ ì‹œê·¸ë‹ˆì²˜: validate_transaction(func: Callable) -> Callable

```python
from typing import Callable, Any

def validate_transaction(func: Callable) -> Callable:
    def wrapper(*args: Any, **kwargs: Any) -> Any:
        # ë°ì½”ë ˆì´í„° í•¨ìˆ˜ êµ¬í˜„
        # amountë¼ëŠ” ì´ë¦„ì˜ ì¸ìë¥¼ ì°¾ìŒ
        if "amount" in kwargs:
            amount = kwargs["amount"]
        elif len(args) > 1:  # self ì œì™¸, ë‘ ë²ˆì§¸ ì¸ìê°€ amountë¼ ê°€ì •
            amount = args[1]
        else:
            raise ValueError("ê±°ë˜ ê¸ˆì•¡(amount)ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

        # ê²€ì¦ ë¡œì§
        if amount <= 0:
            raise ValueError("ê±°ë˜ ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")

        return func(*args, **kwargs)
    return wrapper
```
